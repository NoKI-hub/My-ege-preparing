##### № 27771 Апробация 04.03.26

![alt text](tasks/{BB54596C-5A5F-47A1-8AE6-6FCF2538B658}.png)

```python
from functools import lru_cache

@lru_cache(2030)
def f(n) -> int|None:
    if n == 1: return 1
    if n > 1: return n * f(n - 1)

for i in range(1, 2030):
    f(i)
print((f(2024) - 2 * f(2023)) / f(2022))
```

##### № 27628 Апробация 04.03.26

```python
from functools import lru_cache

@lru_cache(2030)
def f(n):
    if n == 1: return 1
    else: return n * f(n - 1)

[f(i) for i in range(1, 2030)]
print((f(2024) - 5 * f(2023)) / f(2022))
```

Ответ: 4084437
