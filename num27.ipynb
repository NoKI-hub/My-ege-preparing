##### № 25364 ЕГКР 13.12.25 (Уровень: Базовый)

```python
from math import dist


def centroid(clust):
    d = []
    for point in clust:
        d.append((sum(dist(point, i) for i in clust), point))
    return min(d)[1]


clust1 = []
clust2 = []
clust3 = []
cnt = 0
cnt1 = 0
for point in open("27_B_25364.txt"):
    x, y = map(float, point.split())
    if 23 < y < 30:
        clust1.append((x, y))
    if 15 < y < 22:
        clust2.append((x, y))
    if 20 < x < 30:
        clust3.append((x, y))
centr1 = centroid(clust1)
centr2 = centroid(clust2)
centr3 = centroid(clust3)

len_cl1 = len(clust1)
len_cl2 = len(clust2)
len_cl3 = len(clust3)

if len_cl1 > len_cl2:
    mx_cl = clust1
else:
    mx_cl = clust2

if len_cl3 > len(mx_cl):
    mx_cl = clust3

for point in mx_cl:
    if dist(centroid(mx_cl), point) <= 1.2:
        cnt += 1
    if dist(centroid(mx_cl), point) <= 0.75:
        cnt1 += 1

print(cnt, cnt1)
```

##### № 28766 Досрочная волна 2026 (Уровень: Базовый)

![alt text](tasks/{11377EA7-D228-487F-81EB-45E1AEBA123A}.png)

![alt text](tasks/{EF816752-5A3E-4279-BA15-0CE1188DC90B}.png)

```python
from math import dist, floor

def centroid(clust):
    d = []
    for i in clust:
        a = i[:2]
        d.append([sum(dist(a, j[:2]) for j in clust), a])
    return min(d)[1]

def find_stars_by_par(clust, colour, size):
    stars = [star[:2] for star in clust if star[2][0] == colour and star[2][2:] == size]
    return stars

def extremum_dist_between_stars(clust, ext) -> float:
    dists = set()
    for dot1 in clust:
        for dot2 in clust:
            dists.add(dist(dot1, dot2))
    dists = [i for i in dists if i != 0]
    if ext == "max":
        if len(dists) == 0:
            dists = [float("-inf")]
        return max(dists)
    elif ext == "min":
        if len(dists) == 0:
            dists =  [float("inf")]
        return min(dists)
    else:
        return 0
        

def A():
    clust1a = []
    clust2a = []
    clust_r = []

    for line in open("tasks/27_A_28766.txt").readlines():
        x, y = map(float, line.split()[:2])
        par = line.split()[-1]
        if par[2:] == "III" and par[0] == "Y":
            clust_r.append((x, y))
        if 10 <= y <= 14:
            clust1a.append((x, y))
        if 4 <= y <= 8:
            clust2a.append((x, y))

    if len(clust1a) < len(clust2a):
        min_clust = clust1a
    else:
        min_clust = clust2a

    d = []
    centr_min = centroid(min_clust)
    for red in clust_r:
        d.append(dist(red, centr_min))

    print(floor(min(d) * 10000), floor(max(d) * 10000))


def B():
    clust1b = []
    clust2b = []
    clust3b = []

    for line in open("tasks/27_B_28766.txt").readlines():
        x, y = map(float, line.split()[:2])
        par = line.split()[-1]
        if y > 23:
            clust1b.append((x, y, par))
        if y < 23 and x < 20:
            clust2b.append((x, y, par))
        if x > 20:
            clust3b.append((x, y, par))

    y1 = find_stars_by_par(clust1b, "Z", "I")
    y2 = find_stars_by_par(clust2b, "Z", "I")
    y3 = find_stars_by_par(clust3b, "Z", "I")
    
    min_dy1 = extremum_dist_between_stars(y1, "min")
    min_dy2 = extremum_dist_between_stars(y2, "min")
    min_dy3 = extremum_dist_between_stars(y3, "min")
    
    print(floor(min(min_dy1, min_dy2, min_dy3) * 10000), floor(dist(centroid(clust2b), centroid(clust3b)) * 10000))


A()
B()
```
