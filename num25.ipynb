##### № 24165 (Уровень: Средний)

Напишите программу, которая перебирает целые числа, большие 12 365 266, в порядке возрастания и ищет среди них те, которые можно представить в виде произведения ровно пяти различных простых множителей, а сумма этих множителей образует простое число.
В ответе в первом столбце таблицы запишите первые 5 найденных чисел в порядке возрастания, во втором столбце — сумму их простых множителей.

Ответ:


```python
from math import prod
from itertools import combinations
from pprint import pprint


def find_prims(x):
    prims = []
    for i in range(2, x):
        if all(i % d != 0 for d in range(2, int(i ** 0.5) + 1)):
            prims.append(i)
    return prims


x = 12_365_266
prims = find_prims(875000)
result = set()
# print(prims)
while len(result) < 5:
    dels = []
    for delit in prims:
        if x % delit == 0:
            dels.append(delit)
    # print(dels)
    for comb in combinations(dels, r=5):
        comb = list(comb)
        # if len(set(comb)) == len(comb):
        if prod(comb) == x:
            if sum(comb) in prims:
                result.add((x, sum(comb)))
    x += 1

pprint(sorted(result))
```

##### № 24166 (Уровень: Средний)


Напишите программу, которая перебирает целые числа, большие 7 305 678, в порядке возрастания и ищет среди них числа, представленные в виде произведения ровно четырех простых множителей, не обязательно различных, сумма которых является числом палиндромом.
В ответе в первом столбце таблицы запишите первые 5 найденных чисел в порядке возрастания, а во втором столбце - для каждого из чисел соответствующие ему сумму делителей.

Ответы
- 7305747 2002
- 7305818 16561
- 7305986 545
- 7306057 696
- 7306059 606

```python
correct_answer = {
    (7305747, 2002),
    (7305818, 16561),
    (7305986, 545),
    (7306057, 696),
    (7306059, 606)
}


import math
import time
from itertools import product


def prime_digits(to_num: int, from_num: int = 2) -> list[int]:
    assert from_num >= 2, f"from_num should be >= 2, but from_num is {from_num}"

    primes = []
    for possible_prime in range(2, to_num):
        if all(
                possible_prime % d != 0
                for d in range(2, int(possible_prime ** 0.5) + 1)
            ):
            primes.append(possible_prime)
    return primes


def main(x: int) -> set[int]:
    primes: list[int] = prime_digits(to_num=1_125_000)
    result = set()
    while len(result) < 5:
        dels= []
        for delit in primes:
            if x % delit == 0:
                dels.append(delit)
        for comb in product(dels, repeat=4):
            comb = list(comb)
            nsum = str(sum(comb))
            if nsum == nsum[::-1]:
                if x == math.prod(comb):
                    result.add((x, sum(comb)))
        x += 1
    return result



if __name__ == "__main__":
    x = 7_305_678
    start_time = time.perf_counter()

    result = main(x=x)
    assert result == correct_answer, f"Uncorrect answer: {result}; Expected: {correct_answer}"

    runned_time = time.perf_counter() - start_time

    print(f"Result is: {sorted(result)}\n{'# ' + '-' * 100 + ' #'}\nTime to calculate: {runned_time:.3} sec.")
```

##### № 28711 (Уровень: Базовый)

Напишите программу, которая перебирает целые числа, большие 2 400 000, в порядке возрастания и ищет среди них числа, представимые в виде произведения ровно трёх различных простых множителей, каждый из которых содержит в своей записи хотя бы одну цифру 4 или 7. В ответе запишите первые пять чисел в порядке возрастания, справа от каждого числа запишите его наибольший простой делитель.

```python
from math import prod
from itertools import combinations

def is_prime(n):
    if n == 1: return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0: return False
    return True

def find_prims(x) -> list[int]:
    dels = []
    for i in range(2, (x + 100000) // 2 ** 2):
        if is_prime(i):
            if "4" in str(i) or "7" in str(i):
                dels.append(i)
    return dels

def find_max_prime_del(x):
    max_del = 0
    for i in range(x // 2 ** 2, 2, -1):
        if is_prime(i) and x % i == 0:
            max_del = i
            break
    return max_del


cnt = 0
x = 2400000
prims = find_prims(x)
while cnt < 5:
    dels = []
    for delit in prims:
        if x % delit == 0:
            dels.append(delit)
    for var in combinations(dels, r=3):
        if len(set(var)) == len(var):
            if prod(var) == x:
                print(x, find_max_prime_del(x))
                cnt += 1
    x += 1
```
