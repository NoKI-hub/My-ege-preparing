##### № 28922 ЕГКР 18.04.26 (Уровень: Базовый)

![{C447CCA7-5B27-4042-AA0A-017ED7DFB14C}.png](tasks/num1_1.png)

```python
from itertools import permutations

# Данные из таблицы (номера пунктов, соединённых дорогами)
# Каждый элемент списка содержит номера соседних пунктов для пункта с индексом+1
# Например: tab[0] = "47" означает, что пункт 1 соединён с пунктами 4 и 7
table_connections = "47 458 67 125 246 35 138 27".split()

# Данные из графа (буквенные обозначения дорог на схеме)
# Каждая дорога представлена как строка из двух букв (например, "ab" - дорога между A и B)
graph_edges = "ab ah hf fe eg gb cg cb cd df".split()

# Выводим номера пунктов от 1 до 8 для наглядности
print(*range(1, 9))

# Перебираем все возможные перестановки букв A-H
# Каждая перестановка - это вариант сопоставления букв номерам пунктов
for permutation in permutations("abcdefgh"):
    # Проверяем, соответствует ли текущая перестановка данным из таблицы
    # Для каждого ребра графа (буквенная пара) проверяем, есть ли соответствующая дорога в таблице
    is_valid = all(
        # Находим номера пунктов для двух букв ребра (индекс в перестановке + 1)
        str(permutation.index(first_letter) + 1)
        in table_connections[permutation.index(second_letter)]
        for first_letter, second_letter in graph_edges
    )

    # Если перестановка подходит, выводим её
    if is_valid:
        print(*permutation)
```

Ответ: 57

##### № 28749 Досрочная волна 2026 (Уровень: Базовый)

![{7F95CA5D-C022-47BC-8A11-A74034A5F99E}.png](tasks/num1_2.png)

```python
from itertools import permutations

tab = "457 346 24 123 167 257 156".split()
pic = "аб ав бд бв ве дг дк ег ек гк".split()

print(*range(1, 9))

for var in permutations("абвгдек"):
    if all(str(var.index(x) + 1) in tab[var.index(y)] for x, y in pic):
        print(*var)
```

Ответ: 11
