##### № 27764 Апробация 04.03.26 

Откройте файл электронной таблицы, содержащий в каждой строке пять натуральных чисел. Определите количество строк таблицы, для которых выполнены оба условия:
- в строке все числа различны;
- удвоенная сумма максимального и минимального чисел строки равна сумме оставшихся трёх её чисел.

В ответе запишите только число.

```python
cnt = 0
for s in open("tasks/9a.txt").readlines():
    s = [int(x) for x in s.split()]
    print(s)
    max_s = max(s)
    min_s = min(s)
    s_ost = [x for x in s if x != max_s and x != min_s]
    if len(set(s)) == len(s) and 2 * (max_s + min_s) == sum(s_ost):
        cnt += 1
print(cnt)
```

Ответ: 5019

##### № 27621 Апробация 04.03.26

Откройте файл электронной таблицы, содержащей в каждой строке пять натуральных чисел. Определите наименьший номер строки таблицы, для которой выполнены оба условия:
- в строке все числа различны;
- разность максимального и минимального чисел строки равна сумме трёх её оставшихся чисел.

В ответе запишите только число.

```python
cnt = 0
for s in open("tasks/9b.txt").readlines():
    cnt += 1
    s = [int(x) for x in s.split()]
    max_s = max(s)
    min_s = min(s)
    s_ost = [x for x in s if x not in [max_s, min_s]]
    if len(set(s)) == len(s) and max_s - min_s == sum(s_ost):
        print(cnt)
        break
```

Ответ: 1321
