```python
from turtle import *


m = 20
tracer(0)
left(90)
down()

for _ in range(2):
    forward(1 * m)
    left(270)
    forward(16 * m)
    right(90)
up()

back(4 * m)
right(90)
fd(10 * m)
lt(90)
down()

for _ in range(2):
    fd(17 * m)
    rt(90)
    fd(7 * m)
    rt(90)
up()

for x in range(-100, 100):
    for y in range(-100, 100):
        setpos(x * m, y * m)
        dot(5, "blue")
done()
```
