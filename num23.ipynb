##### № 23280 Основная волна 11.06.25 (Уровень: Базовый)

A. Вычесть 1
B. Вычесть 4
C. Найти целую часть от деления на 3
Программа для исполнителя - это последовательность команд.
Сколько существует программ, для которых при исходном числе 19 результатом является число 2, при этом траектория вычислений не содержит числа 8 и содержит 14?

Ответ: 69

```python
def f(x, end):
    if x < end: return 0
    if x == end: return 1
    return f(x - 1, end) + f(x - 4, end) + f(x // 3, end)

print(f(19, 14) * f(14, 2) - f(19, 14) * f(14, 8) * f(8, 2))
```

##### № 24330 (Уровень: Базовый)

Исполнитель преобразует число на экране. У исполнителя есть три команды, которые обозначены латинскими буквами:
A. Вычесть 1
B. Прибавить 3
C. Умножить на 2
Программа для исполнителя – это последовательность команд.
Сколько существует программ, для которых при исходном числе 5 результатом является 100, при этом траектория вычислений не содержит числа кратные 3?

Ответ: 21133

```python
def f(x, end):
    if x % 3 == 0 or x > end + 100: return 0
    if x == end: return 1
    return f(x - 1, end) + f(x + 3, end) + f(x * 2, end)

print(f(5, 100))
```
