##### № 27626 Апробация 04.03.26 

![alt text](tasks/{94A2E016-B62A-4D79-B65C-2E6F1F11A043}.png)

```python
def to_any_ss(x, ss):
    r = ""
    while x > 0:
        r = str(x % ss) + r
        x = x // ss
    return r


min_z = 10000000
for x in range(2031):
    res = 6 ** 2030 + 6 ** 100 - x
    res6 = to_any_ss(res, 6)
    if res6.count("0") < min_z:
        min_z = res6.count("0")
print(min_z)
```

Ответ: 1930

##### № 28935 ЕГКР 18.04.26 (Уровень: Базовый)

![alt text](tasks/{5A8FC039-CAED-4FFA-942E-8852E53C2402}.png)

```python
from string import printable


def value(x: str) -> int:
    return int(f"761{x}035", 23) + int(f"338{x}932", 23)


possible_values = printable[:23]

for x in possible_values:
    val = value(x)
    if val % 22 == 0:
        res = val // 22
        print(res)
        break
```

Ответ: 70045642
